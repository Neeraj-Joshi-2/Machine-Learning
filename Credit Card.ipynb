```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error,accuracy_score
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv('UCI_Credit_Card.csv')
data.head()
```

```python
data.describe()
```

```python
x = data.drop(columns = ['ID' , 'SEX' , 'MARRIAGE' , 'default.payment.next.month'])
y = data['default.payment.next.month']
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x , y , test_size = 0.25)

print('Shape(X_train) : ' , x_train.shape)
print('Shape(y_train) : ' , y_train.shape)
print('Shape(X_test) : ' , x_test .shape)
print('Shape(y_test) : ' , y_test.shape)
```

# RandomForestClassifier

```python
clf = RandomForestClassifier()
```

```python
clf.fit(x_train , y_train)

pred = clf.predict(x_test)

print(pred)

score = clf.score(x_train , y_train)
print('Accuracy : ' , score*100)

score = clf.score(x_test , y_test)
print('Accuracy : ' , score*100)
```

```python
print('Parameters currently in use:\n')
print(clf.get_params())
```

### Classification Report

```python
rep = classification_report(y_test , pred)
print(rep)
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

cf = confusion_matrix(y_test , pred)
```

```python
cf
```

### HyperParameter Tunning

### GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {

    'n_estimators': [10 , 20 , 30],
    'max_depth': [15 , 20 , 30 , 50],
    'criterion': ['gini' , 'entropy']
    
}
```

```python
rc = RandomForestClassifier(random_state = 0)

grid = GridSearchCV(estimator = rc , param_grid = param_grid , cv = 3 , scoring = 'accuracy')

grid.fit(x_train , y_train)
```

```python
print(grid.best_params_)
```

```python
print('Accuracy : ' + str(grid.best_score_))
```

### RandomSearch

```python
from scipy.stats import randint as sp_rand
from random import randrange as r_rand
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = {

    'n_estimators': sp_rand(10 , 100),
    'max_depth': sp_rand(5 , 10),
    'criterion': ['gini' , 'entropy']
    
}

n_itr_search = 20
```

```python
rc = RandomForestClassifier(random_state = 0)

random = RandomizedSearchCV(rc , param_distributions = param_grid , n_iter = n_itr_search , cv = 3 , scoring = 'accuracy')

random.fit(x_train , y_train)
```

```python
print(random.best_params_)
```

```python
print('Accuracy : ' + str(random.best_score_))
```
