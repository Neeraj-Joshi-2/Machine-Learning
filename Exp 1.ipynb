# Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
plt.scatter(df.sepal_length,df.sepal_width)
plt.xlabel('SEPAL LENGTH')
plt.ylabel('SEPAL WIDTH')
plt.title('PLOT')
```

```python
plt.scatter(df.sepal_length,df.sepal_width)
ax = plt.gca()
ax.axis([3,10,2,5])
```

```python
plt.scatter(df.sepal_length,df.sepal_width)
plt.scatter(df.petal_length,df.petal_width)
```

```python
colors = ['green']*(len(df)-1)
colors.append('red')

plt.figure()
plt.scatter(df.petal_length,df.petal_width,c=colors)
```

# Barchart

```python
plt.bar(df.petal_length,df.petal_width,width=0.3, color='r')
```

```python
plt.barh(df.petal_length,df.petal_width,color='b')
```

```python
plt.figure(figsize=(10,5))
plt.bar(df.petal_length,df.petal_width,color='cyan',label='label1')
plt.bar(df.sepal_length,df.sepal_width,color='magenta',label='label2')
plt.legend()
```

# SEABORN

```python
df = pd.read_csv('Rainfalldataset.csv')
df
```

```python
plt.figure(figsize=(8,5))
sns.set_style('darkgrid')
sns.scatterplot(x = df['temp(c)'] , y = df['pressure(mb)'] , marker = '+' , color = 'magenta')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x = 'temp(c)' , y = 'pressure(mb)' , data = df , hue = 'temp(c)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x = 'temp(c)' , y = 'pressure(mb)' , data = df , hue = 'temp(c)' , size = 'temp(c)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.distplot(df['temp(c)'] , color = 'blue')
```

```python
sns.jointplot(df['temp(c)'] , df['pressure(mb)'] , alpha=0.4)
```

```python
sns.jointplot(df['temp(c)'] , df['pressure(mb)'] , kind='hex')
```

```python
sns.set_style('white')
sns.jointplot(df['temp(c)'] , df['pressure(mb)'] , kind='kde', space=0)
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue='species', diag_kind='kde')
```

```python
plt.figure(figsize=(12,8))
plt.subplot(121)
sns.swarmplot('species', 'petal_length', data=iris);
plt.subplot(122)
sns.violinplot('species', 'petal_length', data=iris);
```

# Box and Whisker Plot

```python
plt.figure()
_ = plt.boxplot(df['petal_length'], whis='range')
```

```python
plt.figure()
_ = plt.boxplot(df['sepal_length'], whis='range')
```

```python
import mpl_toolkits.axes_grid1.inset_locator as mpl_il

plt.figure()
plt.boxplot([df['sepal_length'],df['petal_length'] , df['sepal_width']], whis='range')
ax2 = mpl_il.inset_axes(plt.gca(), width='60%', height='40%', loc=2)
ax2.hist(df['sepal_width'], bins=100)
ax2.margins(x=0.5)
```
